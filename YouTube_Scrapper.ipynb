```python
#assignment got updated so removing dislikes column because not needed
data = data.drop('Dislikes', axis=1)
data
```

```python
data["Upload Date"].unique()
```

```python
#replacing "Premiered" word in Upload date values with space
data['Upload Date'] = data['Upload Date'].str.replace('Premiered ', '')
```

```python
# Converting the Upload Date to a date
data['Upload Date'] = pd.to_datetime(data['Upload Date'], format='%b %d, %Y')
data.head()
```

```python
# Filter rows between 'Sep 10, 2019' and 'Sep 10, 2023'
start_date = pd.to_datetime('Sep 10, 2019', format='%b %d, %Y')
end_date = pd.to_datetime('Sep 10, 2023', format='%b %d, %Y')
data = data[(data['Upload Date'] >= start_date) & (data['Upload Date'] <= end_date)]
```

```python
data.info()
```

```python
# Remove commas and ' views' and converting to integer from Views
data['Views'] = data['Views'].str.replace(',', '').str.replace(' views', '').astype(int)
data.head()
```

```python
# Remove 'Comments' and converting to integer from Comments
data['Comments'] = data['Comments'].str.replace(',', '').str.replace(' Comments', '').astype(int)
data.head()
```

```python
data['Likes'].unique()
```

```python
#converting like in integer and removing K by multiplying with 1000
data['Likes'] = data['Likes'].str.replace('K', 'e3').astype(float).astype(int)
data.head()
```

```python
#Calculate the average views count per video for videos uploaded in the last 30 days.(doing on date 20-09-23)
def avg_views(data):
    today = pd.Timestamp.today()
    thirty_days = today - pd.DateOffset(days=30)
    videos = data[data["Upload Date"] >= thirty_days]
    avg = videos["Views"].mean()
    return avg
```

```python
result = avg_views(data)
print("Average is : " ,result)
```

```python
#Identify the video with the highest likes-to-views ratio.
def highest_ratio(data):
    data["Ratio"] = round(data["Likes"]/data["Views"],4)
    max = data["Ratio"].max()
    video = data['Title'][data["Ratio"] == max]
    return video
```

```python
highest_ratio(data)
```

```python
#Find the correlation between the number of likes and the number of comments for the videos.
def correlation(data):
    corr = data["Likes"].corr(data["Comments"])
    return corr
```

```python
corr = correlation(data)
print("Correlation between the number of likes and the number of comments for the videos : ",corr)
```

```python
#Determine the most common day of the week for video uploads.
def common_dayw(data):
    common_day = data['Upload Date'].dt.day_name().mode().iloc[0]
    return common_day
```

```python
common_day = common_dayw(data)
print("Most Common Day of the Week for Video Uploads:", common_day)
```

```python
#Detect any outliers in the views count.
def outliersViews(data):
    Q1 = data["Views"].quantile(0.25)
    Q3 = data["Views"].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data["Views"] < lower_bound) | (data["Views"] > upper_bound)]
    return outliers
```

```python
outliers = outliersViews(data)
print(outliers)
```
